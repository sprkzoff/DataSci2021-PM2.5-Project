```python
import numpy as np
import pandas as pd
import json
```

```python
geo_map = dict()
with open('geo_map.json') as json_file:
    geo_map = json.load(json_file)

rev_geo_map = { v:k for k,v in geo_map.items() }
```

```python
df = pd.read_csv('all_provinces.csv')

all_provinces = ['bangkok','chanthaburi','chiangmai','kanchanaburi','khonkaen','songkhla']

for prv in all_provinces :
    df[prv] = df[['lat', 'long']].apply(lambda x : int(rev_geo_map[prv] == str("%.4f" % x[0].round(4) +',' + "%.4f" % x[1].round(4))) ,axis=1)

df = df.drop(columns=['lat','long'])
df.head(20)
```

```python
df.to_csv('all_provinces_2.csv', index=False)
```
